```python
import pandas as pd 
import csv
from dateutil.parser import parse
import collections
import re
```

```python
def is_date(string, fuzzy=False):
    """
    Return whether the string can be interpreted as a date.
    """
    try: 
        if string.isdigit():
            #print(string)
            return False
        parse(string, fuzzy=fuzzy)
        return True

    except ValueError:
        return False
```

```python
def rewrite_file(new_file_name, rewriten_file_name):
    line=[]
    # open the output file in 'w' mode
    with open(new_file_name, 'w', newline='') as f:

        # create a csv writer object
        writer = csv.writer(f)

        with open(rewriten_file_name, 'r') as csv_file:
            reader = csv.reader(csv_file)

            #skip header
            header = next(reader)

            writer.writerow(header)
            for row in reader:
                #print(row)
                if len(row) > 0:
                    pattern = r'^\d{' + str(4) + r',}$'
                    found = re.findall(pattern,row[0])
                    if found != []:
                        if len(line) > 0:
                            if is_date(line[-1]) == True:
                                new_line = line[:18]+ [' '.join(line[18:len(line)-13])] + line[-13:]
                                writer.writerow(new_line)
                        line = []
                
                line.extend(row)

                
    # explicitly close the files
    csv_file.close()
    f.close()
```

```python
csv.field_size_limit(100000000)
new_file_name = 'test_small.csv'
rewriten_file_name = "E:\Data_exploration\careplan_reports_small.csv"
rewrite_file(new_file_name, rewriten_file_name)
```
